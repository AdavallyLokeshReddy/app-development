# Fit a Linear Regression

```python
#build regression model

import statsmodels.formula.api as smf
model = smf.ols("sunday~daily", data = data1).fit()
```

```python
model.summary()
```

* The predicted value is
* Y (hat) = beta_0 + Beta_1X
* Y (hat) = 13.83556 + 1.3397

- The probability (p-value) for intercept(beta_0) is 0.05
- Thereforethe intercept coefficient may not be that much significant in prediction
- However the p-value for "daily" (beta_1) is 0.00 < 0.05
- Therefore the beta_1 coefficient is highly significant and its contribution to prediction.

```python
x = data1["daily"].values
y = data1["sunday"].values
plt.scatter(x, y, color = "y", marker = "o", s = 30)
b0 = 13.84
b1 = 1.33

#predicted response vector
y_hat = b0 + b1*x

#plotting the regression line
plt.plot(x, y_hat, color = "b")

#putting labels
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
import statsmodels.formula.api as smf
model1 = smf.ols("sunday ~ daily", data=data1).fit()
```

```python
model1.params
```

```python
print(f'model t-values:\n{model1.tvalues}\n----------------\nmodel p-values: \n{model1.pvalues}')

```

```python
(model1.rsquared,model1.rsquared_adj)
```

Predict new data point

```python
newdata=pd.Series([200,300,1500])
```

```python
data_pred = pd.DataFrame(newdata, columns=['daily'])
data_pred
```

```python
model1.predict(data_pred)
```

```python
pred = model1.predict(data1["daily"])
pred
```

```python
data1["Y_hat"] = pred
data1
```

```python
data1["residuals"]=data1["sunday"]-data1["Y_hat"]
data1
```

```python
mse = np.mean((data1["daily"]-data1["Y_hat"])**2)
rmse = np.sqrt(mse)
print("MSE: ",mse)
print("RMSE: ",rmse)
```

```python
plt.scatter(data1["Y_hat"], data1["residuals"])
```

#### observation:
the residual datapoints are randomly scattered
hence the assumption of homoscedasticity is satisfied (constant variance in residuals)

```python
import statsmodels.api as sm
sm.qqplot(data1["residuals"], line = '45', fit=True)
plt.show()
```

```python
sns.histplot(data1["residuals"], kde = True)
```

```python
#### Observations
the data points are seen to closely follow the reference line of normality
hence the residuals are approx normally distributed as also can be seen from the histogram 
```

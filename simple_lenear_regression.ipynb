```python
data1 = pd.read_csv('NewspaperData.csv')
print(data1.columns)
```

```python
sns.boxplot(data1['daily'])
plt.title('Boxplot of daily')
plt.xlabel('daily')
plt.show()
```

```python
sns.histplot(data1['daily'], kde=True)
plt.title('Histogram of daily')
plt.xlabel('daily')
plt.ylabel('Frequency')
plt.show()
```

#### it is right skeweed

```python
plt.kde(data1['daily'])
plt.title('KDE Plot of daily')
plt.xlabel('daily')
plt.ylabel('Density')
plt.show()
```

```python
plt.scatter(data1['daily'], data1['sunday'])
plt.title('Scatter Plot of daily vs Sunday')
plt.xlabel('daily')
plt.ylabel('Sunday')
plt.show()
```

```python
model = smf.ols("sunday~daily",data = data1).fit()
```

```python
model.summary()
```

```python
x = data1["daily"].values
y = data1["sunday"].values
plt.scatter(x, y, color = "m", marker = "0", s = 30)
b0 = 13.84
b1 = 1.33
y_hat = b0 + b1*x
plt.plot(x, y_hat, color = "g")
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
